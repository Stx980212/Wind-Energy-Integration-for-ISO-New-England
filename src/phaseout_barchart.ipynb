```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('phaseout_cf.csv')
data
```

```python
x=data.iloc[:,0].values
y = data.iloc[:,1].values
x = ["13 \n High D, High CF","22 \n Low D, Medium CF","33 \n High D, Low CF","61 \n Low D, Low CF","64 \n High D, Medium CF","66 \n Medium D, Medium CF","81 \n Low D, High CF"]
```

```python
fig = plt.figure(figsize=(6,4),dpi=800)

ax = fig.add_subplot(111)
plt.title("Wind Capacity Needed for Phasing Out Coal Generators")
ax.set_xlabel("Scenarios")
ax.set_ylabel(r"Wind Capacity (MWh)")
plt.bar(x,y,data=data,width = 0.5)
plt.xticks(rotation=10,size = 7)
plt.grid(alpha = 0.5)
plt.savefig("Wind Capacity Needed for Phasing Out Coal Generators.jpg", dpi = 800)
```
